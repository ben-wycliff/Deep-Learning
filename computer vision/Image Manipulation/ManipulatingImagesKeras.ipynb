# Importing Modules

```python
from keras.preprocessing.image import load_img, img_to_array, array_to_img, save_img
from matplotlib import pyplot
```

# Loading image

```python
img = load_img("image.jpg")

print(f"Type {type(img)}")
print(f"Format {img.format}")
print(f"Size {img.size}")
print(f"Mode {img.mode}")
pyplot.imshow(img)
pyplot.show()
```

# Converting From Image To Array

```python
img_array = img_to_array(img)
print(img_array.dtype)
print(img_array.shape)
```

# Converting Back To PIL Image

```python
img_pil = array_to_img(img_array)
print(type(img_pil))
pyplot.imshow(img_pil)
pyplot.show()
```

# Converts To Grayscale, Saves Image

```python
img_gray = load_img("hogkong.jpg", color_mode="grayscale")
img_gray_array = img_to_array(img_gray)
save_img("hogkonggray.jpg", img_gray_array)
image2 = load_img("hogkonggray.jpg")
print(type(image2.format))
print(image2.format)
print(image2.mode)
print(image2.size)
pyplot.imshow(image2)
```

# Scaling Image Pixel Data With Keras

```python
from keras.datasets import mnist
```

### Exploring Dataset

```python
(trainX, trainY), (testX, testY) = mnist.load_data()
print("Train: ", trainX.shape, trainY.shape)
print("Test: ", testX.shape, testY.shape)
print("Train: ", trainX.min(), trainX.max(), trainX.mean(), trainX.std())
print("Test: ", testX.min(), testX.max(), testX.mean(), testX.std())
```
